```python
import numpy as np
import pandas as pd
```

### Importing Graph Plotting library

```python
import matplotlib.pyplot as plt
```

### Plotting TIme taken by different algorithms

```python
plt.figure(figsize=(10,5))
plt.bar(data.Algorithms,data.Time_Taken)
plt.grid()
plt.ylabel('Time Taken', fontsize=20)
plt.xlabel('Machine Learning Algorithms', fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

### Plotting Training Accuracy by different algorithms

```python
plt.figure(figsize=(10,5))
plt.bar(data.Algorithms,data.Training_Accuracy)
plt.grid()
plt.ylabel('Training_Accuracy', fontsize=20)
plt.xlabel('Machine Learning Algorithms', fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

### Plotting Testing Accuracy by different algorithms

```python
plt.figure(figsize=(10,5))
plt.bar(data.Algorithms,data.Testing_Accuracy)
plt.grid()
plt.ylabel('Testing_Accuracy', fontsize=20)
plt.xlabel('Machine Learning Algorithms', fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
